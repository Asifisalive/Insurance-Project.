```python
#importing the packages
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/insurance.csv")
```

```python
data.head()
```

Data Cleaning

```python
#converting categorical variables into numerical format - one hot coding
data['sex']=data['sex'].apply({'female':1,'male':0}.get)
```

```python
data.head()
```

```python
data['smoker']=data['smoker'].apply({'yes':1,'no':0}.get)
data['region']=data['region'].apply({'southeast':1,'southwest':2,'northeast':3,'northwest':4}.get)
```

```python
data.head()
```

dividing the data set into training and testing part

```python
x=data[['age','sex','bmi','children','smoker','region']] #indepndent variables
y=data[['expenses']] #dependent varibales
```

splitting the data into taringning and testing set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8) #80% data used for trainig (learning)
```

creating our SMLM using linear Rregression Algorithm

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression() #Creating Algorithm
```

```python
regression.fit(x_train,y_train) #.fit () used for training
```

```python
y_pred=regression.predict(x_test)
```

```python
print(y_pred)
```

```python
#accuracy of SMLM
regression.score(x,y)
```
